# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area = [1.2,2.4,3.6,2.5,1.8,3.9,5.8,4.2,4.1,3.7,6.7,5.8,4.9,7.2,3.5,2.6,4.6,5.8,1.3,3.3]
price= [180,220,290,280,240,420,480,380,360,390,590,620,540,670,410,320,450,490,180,380]
```

```python
m = 0.9
c = 0.8
x = np.array(area)
y = np.array(price)
Lr = 0.002
for i in range(5):
    yhat = m*x + c
    e = np.mean((y-yhat)**2)
    # gradient descent
    dedm = 2*np.dot((yhat-y),x)
    m = m - Lr*dedm
    dedc = 2*np.dot((yhat-y),np.ones(len(x)))
    c = c - Lr*dedc 
    print(e)
    
    plt.plot(x,yhat,c='r')
    plt.scatter(x,y)
    plt.show()
```
